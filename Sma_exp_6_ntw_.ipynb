```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.Graph()
```

```python
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])
```

```python
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 1), (4, 6), (7, 1), (3, 8), (2, 6), (1, 8)])
```

```python
new_nodes = range(9, 29)
new_edges = [(9, 1), (10, 2), (11, 3), (12, 4), (13, 5), (14, 6), (15, 7), (16, 8),
             (17, 1), (18, 2), (19, 3), (20, 4), (21, 5), (22, 6), (23, 7), (24, 8),
             (25, 1), (26, 2), (27, 3), (28, 4)]
G.add_nodes_from(new_nodes)
G.add_edges_from(new_edges)
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
cliques = list(nx.find_cliques(G))
print("Cliques:", cliques)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=12, font_weight='bold')
for clique in cliques:
    nx.draw_networkx_nodes(G, pos, nodelist=clique, node_color='red', node_size=800)  # Highlight cliques in red
plt.title('Graph with Cliques Highlighted')
plt.show()
```

```python
betweenness_centrality = nx.betweenness_centrality(G)
print("Betweenness centrality:", betweenness_centrality)
```

```python
density = nx.density(G)
print("Density:", density)
```

```python
bridges = list(nx.bridges(G))
print("Bridges:", bridges)
```

```python
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=800, font_size=12, font_weight='bold')
nx.draw_networkx_edges(G, pos, edgelist=bridges, edge_color='red', width=2)  # Highlight bridges in red
plt.title('Graph with Bridges Highlighted')
plt.show()
```

```python
ego_net = nx.ego_graph(G, 1)
# Visualize the egocentric network
nx.draw(ego_net, with_labels=True)
plt.show()
```
